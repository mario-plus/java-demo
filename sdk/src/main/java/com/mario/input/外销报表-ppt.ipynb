# 毛利率

```python
data22_table毛利率 = data22.pivot_table(index='【透】行业类型',
                                  columns=['会计年度','【透】期间'],
                                  values='【透】毛利率',
                                  aggfunc='sum',
                                  fill_value=0)

data22_table毛利率['合计'] = data22_table毛['合计']/data22_table['合计']
```

```python
data22_table毛利率.loc['毛利率'] =  data22_table毛.loc['毛利'] / data22_table.loc['销售收入']
data22_table毛利率.loc['专显'] = data22_table毛.loc['专显'] / data22_table.loc['专显']
data22_table毛利率.loc['其他'] = data22_table毛.loc['其他'] / data22_table.loc['其他']
data22_table毛利率.loc['户内商显'] = data22_table毛.loc['户内商显'] / data22_table.loc['户内商显']
data22_table毛利率.loc['户外商显'] = data22_table毛.loc['户外商显'] / data22_table.loc['户外商显']
data22_table毛利率.loc['文体'] = data22_table毛.loc['文体'] / data22_table.loc['文体'].replace(0, 1)
data22_table毛利率.loc['照明'] = data22_table毛.loc['照明'] / data22_table.loc['照明']                             
data22_table毛利率.loc['租赁'] = data22_table毛.loc['租赁'] / data22_table.loc['租赁']                             
data22_table毛利率.loc['蓝普'] = data22_table毛.loc['蓝普'] / data22_table.loc['蓝普']
data22_table毛利率.loc['蓝普分销'] = data22_table毛.loc['蓝普分销'] / data22_table.loc['蓝普分销']
```

```python
def fill_func(x):
    if x > 1:
        return 0
    else:
        return x

# Apply the function to all elements in the DataFrame
data22_table毛利率 = data22_table毛利率.applymap(fill_func)
```

```python
concatenated_table = pd.concat([data22_table, data22_table毛,data22_table毛利率])
```

# 区域&战区

```python
区域战区 = 国际收毛[国际收毛['【透】期间'].isin([1, 2]) & (国际收毛['会计年度'] == 2024) & (国际收毛['是否关联交易'] == '非关联交易')]
区域战区 = 区域战区[['会计年度','【透】行业类型','【透】收入', '【透】成本', '【透】毛利', '【透】区域','【透】毛利率']]
```

```python
区域战区['【透】行业类型'] = 区域战区['【透】行业类型'].apply(seg)
```

```python
分区收入 = 区域战区.pivot_table(index='【透】行业类型',
                                  columns='【透】区域',
                                  values='【透】收入',
                                  aggfunc='sum',
                                  fill_value=0) 

分区收入 = 分区收入.apply(lambda x: x / 10000).applymap(lambda x: round(x, 2))
分区收入.loc['销售收入'] = 分区收入.sum(axis=0)
分区收入['合计'] = 分区收入.sum(axis=1)


分区收入.loc[len(分区收入)] = [None] * len(分区收入.columns)

分区毛利 = 区域战区.pivot_table(index='【透】行业类型',
                                  columns='【透】区域',
                                  values='【透】毛利',
                                  aggfunc='sum',
                                  fill_value=0)
分区毛利 = 分区毛利.apply(lambda x: x / 10000).applymap(lambda x: round(x, 2))
分区毛利.loc['毛利'] = 分区毛利.sum(axis=0)
分区毛利['合计'] = 分区毛利.sum(axis=1)
分区毛利.loc[len(分区毛利)] = [None] * len(分区毛利.columns)
```

```python
分区毛利率 = 区域战区.pivot_table(index='【透】行业类型',
                                  columns='【透】区域',
                                  values='【透】毛利率',
                                  aggfunc='sum',
                                  fill_value=0)

分区毛利率['合计'] = 分区毛利['合计']/分区收入['合计']
```
