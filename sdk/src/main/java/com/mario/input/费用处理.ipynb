# 中东非&俄语区

```python
# 初始化一个空的DataFrame来存储结果
new_rows = []

# 遍历原始DataFrame
for index, row in 费用表.iterrows():
    if row['部门描述2'] == '中东非&俄语区':
        # 计算拆分的金额
        amt = row['本位币金额']
        amt_russian = round(amt * 0.117, 2)  # 俄语区金额，保留两位小数
        amt_middle_east = round(amt * 0.883, 2)  # 中东非金额，保留两位小数
        
        # 创建对应中东非和俄语区的行，并添加到new_rows列表中
        row_middle_east = row.copy()
        row_middle_east['部门描述2'] = '中东非'
        row_middle_east['本位币金额'] = amt_middle_east
        new_rows.append(row_middle_east)
        
        row_russian = row.copy()
        row_russian['部门描述2'] = '俄语区'
        row_russian['本位币金额'] = amt_russian
        new_rows.append(row_russian)
    else:
        # 如果不需要拆分，则直接将行添加到new_rows列表中
        new_rows.append(row)

# 使用pd.DataFrame()将列表转换回DataFrame
new_df = pd.DataFrame(new_rows)

# 由于直接使用列表来添加数据，每行数据都会被当作一个Series对象处理，
# 这可能会导致最终DataFrame的索引和列名的数据类型与预期不同，
# 可以使用reset_index()方法重置索引，并使用rename()方法确保列名保持一致。
new_df = new_df.reset_index(drop=True)
```
